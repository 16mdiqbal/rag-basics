```python
from langchain_community.document_loaders import DirectoryLoader, TextLoader

docs = DirectoryLoader(
    "../docs/text_file", 
    glob="*.txt", 
    loader_cls=TextLoader,
    show_progress=True
).load()

print(f"📁 Loaded {len(docs)} documents.")
```

```python
for i, doc in enumerate(docs):
    print(f"Loading document {i+1} \n Metadata: {doc.metadata['source']} \n Content : {doc.page_content[:100]}")
    print("-" * 80)
```

## Document Splitter

```python
from langchain_text_splitters import (
    RecursiveCharacterTextSplitter,
    CharacterTextSplitter,
    TokenTextSplitter,
)
```

# CharacterTextSplitter

```python
text = docs[0].page_content
text
```

```python
character_splitter = CharacterTextSplitter(
    separator="\n\n",
    chunk_size=500,
    chunk_overlap=50,
    length_function=len
)
character_chunks = character_splitter.split_text(text)
```

```python
print(f"Number of character chunks: {len(character_chunks)}")
#print(f"First character chunk: {character_chunks[0]}")


for i, chunk in enumerate(character_chunks):
    print(f"Chunk {i+1} : {chunk}")
    print("-" * 80)
```
